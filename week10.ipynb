```python
encode = input()

def encoding(encode): 
    codenum = ""
    for i in range(len(encode)):
        if encode[i] != " ":
            r = chr(ord(encode[i]) + 1)
        else:
            r = " "
        codenum += r

    start = codenum[0:3]
    end = codenum[4:7]
    codenum = end + " " + start
        
    return codenum
codenum = encoding(encode)
print("encoding :",codenum)

decode = input()

def decoding(codenum):
    decode =""
    start = codenum[0:3]
    end = codenum[4:7]
    codenum = end + " " + start
    
    for i in range(len(codenum)):
        if codenum[i] != " ":
            r = chr(ord(codenum[i]) - 1)
        else:
            r = " "
        decode += r
        
    return decode


codenum2 = decoding(codenum)
print("decoding :",codenum2)
```
